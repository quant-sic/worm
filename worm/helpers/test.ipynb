```python
from worm.helpers.mu import grf
import numpy as np
import functools
def gaussian_2d(x, y, sigma):
    return np.exp(-((x**2 + y**2) / (2 * sigma**2))) * (1 / (2 * np.pi * sigma**2))

L=100
out = grf((L, L), functools.partial(gaussian_2d, sigma=10))

import matplotlib.pyplot as plt




def get_offset_rescaled_trapping_potential(potential:np.ndarray, desired_abs_max:float):
    abs_max = abs(potential).max()
    return potential * desired_abs_max / abs_max

plt.imshow(get_offset_rescaled_trapping_potential(out, 1.0))
plt.show()
plt.plot(out[0, :])
```

```python
import FyeldGenerator
import numpy as np


def Pkgen(n):
    def Pk(k):
        return np.power(k, -n)

    return Pk

def distrib(shape):
    # Build a unit-distribution of complex numbers with random phase
    a = np.random.normal(loc=0, scale=1, size=shape)
    b = np.random.normal(loc=0, scale=1, size=shape)
    return a + 1j * b

shape = (100, 100)

field = FyeldGenerator.generate_field(distrib, Pkgen(30*10**(-1)), shape)

import matplotlib.pyplot as plt
plt.imshow(field)
```
